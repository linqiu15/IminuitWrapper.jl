```python
from iminuit import Minuit
```

```python
import numpy as np
```

```python
def fcn(x):
    return np.sum([xi**2 for xi in x])
```

```python
def gcn(x):
    return [2*xi for xi in x]
```

```python
model=Minuit(fcn,[3,4,5],name=["x","y","z"],grad=gcn)
```

```python
model.migrad()
```

```python
model.minos("x","y",cl=0.68)
```

```python
model.mncontour("x","y",cl=0.68)
```

```python
model.draw_mncontour("x", "y", cl=(0.68, 0.9, 0.99));
```

```python
model.mnprofile("x","y",size=40,bound=[-10,10])
```
